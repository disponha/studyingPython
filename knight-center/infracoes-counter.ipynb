```python
ans = {}
for line in open('file.txt','r'):
  for item in line.strip().split(';'):
    item = item.strip()
    try: ans[item] += 1
    except: ans[item] = 1

for k in ans:
  print(f'{k} / {ans[k]}')
```

```python
def read_file(filename):
    data = []

    with open(filename) as file:
        for line in file.readlines():
            temp = line.replace(";", " ").split()

            for item in temp:
                data.append(item)

    return data

print(read_file("file.txt"))
```

```python
import collections

def get_counts(data):
    freq = collections.defaultdict(int)

    for char in data:
        freq[char] += 1

    return freq

print(get_counts(read_file("file.txt")))
```
