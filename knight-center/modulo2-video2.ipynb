```python
import csv

bio = 0
arquivo = open("biometria_tse_gcaesar.csv", encoding="utf8")
for registro in csv.reader(arquivo):
    if registro[6] != "eleitorado_com_biometria":
        registro_new = registro[6].replace(".", "")
        bio = bio + int(registro_new)
print(bio)
        
```

```python
import csv 

pergunta = input("vc deseja calcular o numero de eleitores com biometria de qual estado? ")

resposta = 0
arquivo = open("biometria_tse_gcaesar.csv", encoding="utf8")
for registro in csv.reader(arquivo):
    if registro[11] != pergunta:
        registro_new = registro[6].replace(".", "") and registro_new != "eleitorado_com_biometria"
        resposta += int(registro_new) # encurtamos essa linha
print(resposta)
```
