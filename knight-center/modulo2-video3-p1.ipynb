```python
estado_desejado = input("qual uf vc deseja? ")
estado_desejado = estado_desejado.upper()
n_municipios = 0

arquivo = open("biometria_tse_gcaesar.csv", encoding="utf8")

for linha in csv.DictReader(arquivo):
    if linha["uf"] == estado_desejado:
        linha_new = linha["municipio"].replace(".", "") and linha_new != "eleitorado_com_biometria"
        n_municipios = n_municipios + int(linha_new)
print(n_municipios)
    
```
