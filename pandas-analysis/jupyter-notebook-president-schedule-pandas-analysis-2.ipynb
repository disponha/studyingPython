```python
import pandas as pd
```

```python
agenda2016 = pd.read_csv("agenda2016.csv")
```

```python
agenda2016.head()
```

```python
agenda2016.columns = ["quando", "oque", "onde", "nda"]
agenda2016.head(2)
```

```python
agenda2016[["ano", "mes", "dia"]] = agenda2016.quando.str.split("-", expand=True)
```

```python
agenda2016.head(2)
```

```python
agenda2016[["dia", "hora"]] = agenda2016.dia.str.split(" ", expand=True)
agenda2016.head(2)
```

```python
agenda2016_limpa = agenda2016.drop(["quando", "nda"], axis=1)
```

```python
agenda2016_limpa.head(2)
```

```python
agenda2016_limpa[["onde"]] = agenda2016_limpa.onde.str.split("Local: ", expand=True)
agenda2016_limpa.head(2)
```

```python
agenda2016_limpa2 = agenda2016_limpa.drop(["nda"], axis=1)
```

```python
agenda2016_limpa2.head(15)
```

```python
agenda2016_limpa_final = agenda2016_limpa2.fillna("semLocal") # corrigindo o nome da variavel
```

```python
agenda2016_limpa_final.head(10)
```

```python
import re

list = []

s = agenda2016_limpa_final["oque"][0] ## se eu mudar esse numero, eu mudo a linha 

# Recreate dataframe with string (s) two times
df = pd.DataFrame(dict(oque=[s,s]))

titles = ['Deputado Federal ', 'General ', 'Ex-presidente ', 'Senadora ', 'Senador ', 'do Exército', 'Tenente-Brigadeiro']

df['oque'] = (df['oque'].apply(lambda x: 
              [re.sub('|'.join(titles),'',i.split(', ')[0]) for i in x.split('; ')]))

# Append first row and print
list.append(df['oque'][0])
print(list)
```
