```python
agenda2018 = pd.read_csv("agenda2018.csv")
```

```python
agenda2018.columns = ["quando", "oque", "onde", "nda"]
agenda2018.head(2)
```

```python
agenda2018[["ano", "mes", "dia"]] = agenda2018.quando.str.split("-", expand=True)
```

```python
agenda2018.head(2)
```

```python
agenda2018[["dia", "hora"]] = agenda2018.dia.str.split(" ", expand=True)
```

```python
agenda2018.head(2)
```

```python
agenda2018_limpa = agenda2018.drop(["quando", "nda"], axis=1)
```

```python
agenda2018_limpa.head(2)
```

```python
agenda2018_limpa[["nda", "onde"]] = agenda2018_limpa.onde.str.split("Local: ", expand=True)
```

```python
agenda2018_limpa.head(2)
```

```python
agenda2018_limpaFinal = agenda2018_limpa.drop(["nda"], axis=1)
```

```python
agenda2018_limpaFinal.head(15)
```

```python
agenda2016 = agenda2018_limpaFinal.fillna("semLocal") # corrigindo o nome da variavel
```

```python
agenda2016.head()
```

```python
encontros2016 = agenda2016["oque"].str.lower().str.split(",")
```
